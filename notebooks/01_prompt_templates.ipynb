```python
from langchain_ollama import OllamaLLM
from langchain_core.prompts import PromptTemplate
```

```python
llm = OllamaLLM(model="llama3.2")

# Without prompt template - naive way
topic = "neural networks"
audience = " a 10 year old"
raw_prompt = f"Explain {topic} to {audience} in 3 sentences."
print("Raw prompt:", raw_prompt)
print("\nResponse:", llm.invoke(raw_prompt))

print("\n" + "─"*60 + "\n")


#With prompt template - langchain way
template = PromptTemplate(
    input_variables=["topic", "audience"],
    template = "Explain {topic} to {audience} in 3 sentences"
)

#Reuse the same template 
for topic, audience in [
    ("neural networks",    "a 10 year old"),
    ("gradient descent",   "a cricket commentator"),
    ("transformer models", "a retired banker"),
]:
    prompt = template.format(topic=topic, audience=audience)
    response = llm.invoke(prompt)
    print(f"Topic: {topic} | Audience: {audience}")
    print(f"response: {response}\n")
```
